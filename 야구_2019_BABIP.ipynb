```python
baseball_2019_bat = pd.read_csv('./data/19년_타자_기록.csv', sep='\t', encoding='utf-8')
```

```python
baseball_2019_pit = pd.read_csv('./data/19년_투수_기록.csv', sep='\t', encoding='utf-8')
```

```python
baseball_2019_bat[['BABIP']] = baseball_2019_bat[['BABIP']].apply(pd.to_numeric, errors='coerce')
```

```python
baseball_2019_pit[['BABIP']] = baseball_2019_pit[['BABIP']].apply(pd.to_numeric, errors='coerce')
```

```python
baseball_2019_bat.info()
```

```python
baseball_2019_bat.sort_values(by='팀명',ascending=False).head()
```

```python
baseball_2019_pit.sort_values(by='팀명',ascending=False).head()
```

```python
bat_2019= baseball_2019_bat.groupby('팀명')['BABIP'].mean().sort_values(ascending=False)
```

```python
pit_2019 = baseball_2019_pit.groupby('팀명')['BABIP'].mean().sort_values(ascending=False)
```

```python
sum_2019 = (bat_2019 + pit_2019)/2

sum_2019.sort_values(ascending=False)
```
